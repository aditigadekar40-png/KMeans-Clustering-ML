**Step 1: Import Libraries**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
```

**Step 2: Load and prepare data**

```python
age=[23,25,28,35,37,40,34,41,39,38]
hours= [3,2,2,4,5,4,3,5,4,3]
print("Age:", age)
print("hours:", hours)
data = np.column_stack((age, hours))
print ("data",data)
#Number of clusters
k = 2
```

```python
# Initialize and fit the K-Means model
kmeans = KMeans(n_clusters=k, random_state=0, n_init=2) # n_init number of iterartions
kmeans.fit(data)
```

```python
# Get cluster labels and centroids
labels = kmeans.labels_
centroids = kmeans.cluster_centers_
# Print the results
print("Cluster Labels:", labels)
print("Cluster Centroids:\n", centroids)
```

```python
# Visualize the clusters
plt.figure(figsize=(8, 6))
plt.scatter(data[:, 0], data[:, 1], c=labels, cmap='viridis', s=50, label='Data Points')
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', marker='X', s=200, label='Centroids')
plt.title('K-Means Clustering (k=2)')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.legend()
plt.grid(True)
plt.show()
```
